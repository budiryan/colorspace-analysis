```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
urine = pd.read_excel("data plot Final analysis.xlsx", skiprows=1)
```

```python
idx = 84
```

```python
jeffry_urine_time = urine['Time'].iloc[:idx + 1].dropna()
jeffry_urine_day = urine['Day'].iloc[:idx + 1].dropna()
jeffry_urine_avg = urine['Average'].iloc[:idx + 1].dropna()
```

```python
jeffry = pd.DataFrame({'day': jeffry_urine_day, 'time': jeffry_urine_time, 'average': jeffry_urine_avg})
```

```python
jeffry['time'] = pd.to_datetime(jeffry['time'])
```

```python
urine.columns
```

```python
budi_urine_time = urine['Time.1'].dropna()
budi_urine_day = urine['Day '].dropna()
budi_urine_avg = urine['Average.1'].dropna()
```

```python
budi_urine_time = pd.to_datetime(budi_urine_time)
```

```python
budi = pd.DataFrame({'day': budi_urine_day, 'time': budi_urine_time, 'average': budi_urine_avg})
```

```python
plt.figure(figsize=(18,6), dpi=60)
plt.xlabel('Time (Year-Month-Date)', fontsize=15)
plt.ylabel('Average Urine Color Level (8 is most yellow)', fontsize=15)
plt.title('Average Urine Color Timeseries', fontsize=25)
line1, = plt.plot(jeffry['time'], jeffry['average'])
line2, = plt.plot(budi['time'], budi['average'])
plt.legend([line1, line2], ["Person 1","Person 2"], loc=2, fontsize=25)
plt.show()
```
