```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib as mpl
import matplotlib.patches as mpatches

red = mpatches.Patch(color='red', label='Red')
green = mpatches.Patch(color='green', label='Green')
blue = mpatches.Patch(color='blue', label='Blue')

l = mpatches.Patch(color='red', label='Luminence')
a = mpatches.Patch(color='green', label='A*')
b = mpatches.Patch(color='blue', label='B*')

mpl.rcParams['axes.prop_cycle'] = mpl.cycler(color=["r", "g", "b"]) 
```

```python
rgb_lights = pd.read_csv('rgb-lights.csv')
```

```python
rgb_no_lights = pd.read_csv('rgb-no-lights.csv')
```

```python
lab_lights = pd.read_csv('lab-lights.csv')
```

```python
lab_no_lights = pd.read_csv('lab-no-lights.csv')
```

```python
plt.xlabel('Urine Color Level')
plt.ylabel('Color Value (RGB)')
plt.title('RGB Urine Color Chart')

line1, = plt.plot(rgb_lights['Level'], rgb_lights['R'])
line2, = plt.plot(rgb_lights['Level'], rgb_lights['G'])
line3, = plt.plot(rgb_lights['Level'], rgb_lights['B'])
line4, = plt.plot(rgb_lights['Level'], rgb_no_lights['R'], linestyle='--')
line5, = plt.plot(rgb_lights['Level'], rgb_no_lights['G'], linestyle='--')
line6, = plt.plot(rgb_lights['Level'], rgb_no_lights['B'], linestyle='--')

# Create a legend for the first line.
first_legend = plt.legend([line1, line2, line3], ['Red', 'Green', 'Blue'], loc=3)

# Add the legend manually to the current Axes.
ax = plt.gca().add_artist(first_legend)

# Create another legend for the second line.
legends = plt.legend([line5, line6], ['With Lighting', 'Without Lighting'], loc=(0.015,0.25))
count = 0
for leg in legends.legendHandles:
    leg.set_color('black')
    if(count == 0):
        leg.set_linestyle('-')
    count += 1
plt.show()
```

```python
plt.xlabel('Urine Color Level')
plt.ylabel('Color Value (CIELAB)')
plt.title('CIELAB Urine Color Chart')

line1, = plt.plot(lab_lights['Level'], lab_lights['L'])
line2, = plt.plot(lab_lights['Level'], lab_lights['A'])
line3, = plt.plot(lab_lights['Level'], lab_lights['B'])
line4, = plt.plot(lab_lights['Level'], lab_no_lights['L'], linestyle='--')
line5, = plt.plot(lab_lights['Level'], lab_no_lights['A'], linestyle='--')
line6, = plt.plot(lab_lights['Level'], lab_no_lights['B'], linestyle='--')
# Create a legend for the first line.
first_legend = plt.legend([line1, line2, line3], ['Luminence', 'A*', 'B*'])

# Add the legend manually to the current Axes.
ax = plt.gca().add_artist(first_legend)

# Create another legend for the second line.
legends = plt.legend([line5, line6], ['With Lighting', 'Without Lighting'], loc=(0.015,0.62))
count = 0
for leg in legends.legendHandles:
    leg.set_color('black')
    if(count == 0):
        leg.set_linestyle('-')
    count += 1
plt.show()
```
