```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
night = pd.read_excel("night.xlsx")
```

```python
morning = pd.read_excel("morning.xlsx")
```

```python
plt.figure(figsize=(18,6), dpi=60)
plt.xlabel('Time', fontsize=15)
plt.ylabel('Urine Color Level(8 is the most yellow)', fontsize=15)
plt.title('Night vs Morning Urine Color Timeseries', fontsize=25)
threshold = np.zeros(night['Time'].shape[0])
threshold += 3.5
threshold = pd.Series(threshold)
line1, = plt.plot(night['Time'], night['Average'], linestyle='--', color='r')
line2, = plt.plot(morning['Time'], morning['Average'], color='r')
line3, = plt.plot(night['Time'], threshold, color='orange')
_ = plt.legend([line1, line2], ["Night","Morning"], loc=2, fontsize=25)
```
